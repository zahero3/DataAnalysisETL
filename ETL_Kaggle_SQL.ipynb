# Transform Data

```python
#now, getting unzipped csv dataset to python pandas and performing some cleaning
import pandas as pd

#na_values will make the values present in dataset as nan
df = pd.read_csv('orders.csv',na_values=['Not Available', 'unknown'])
#df.head(20)

df['Ship Mode'].unique()
```

```python
#renaming column names as per standard format
df.columns
df.rename(columns={'Order Id':'order_id','City':'city'})

#changing all column name to lower case
df.columns = df.columns.str.lower()

#replacing spaces in col names with underscore
df.columns = df.columns.str.replace(' ','_')
df.head(5)
```

```python
#Deriving new columns, discount, salePrice, profit

df['discount'] = (df['list_price'] * df['discount_percent'])/100
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df.head(10)
```

```python
#converting data type of date time

df.dtypes  #shows data type of each column in df

df['order_date'] = pd.to_datetime(df['order_date'],format='%Y-%m-%d')
df.dtypes
```

```python
#Dropping not required columns from DF

df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df.head(5)
```
